```python
import pandas as pd

train = pd.read_csv('../input/feedback-prize-english-language-learning/train.csv')
test = pd.read_csv('../input/feedback-prize-english-language-learning/test.csv')
submission = pd.read_csv('../input/feedback-prize-english-language-learning/sample_submission.csv')

test_text = test.full_text
text = train.full_text
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.tree import DecisionTreeClassifier

count = CountVectorizer()
tfidf = TfidfTransformer(use_idf=True, norm='l2', smooth_idf=True)

y1 = train['cohesion'].astype(int)
y2 = train['syntax'].astype(int)
y3 = train['vocabulary'].astype(int)
y4 = train['phraseology'].astype(int)
y5 = train['grammar'].astype(int)
y6 = train['conventions'].astype(int)

y1 = y1*2-1
y2 = y2*2-1
y3 = y3*2-1
y4 = y4*2-1
y5 = y5*2-1
y6 = y6*2-1

```

```python
print(text.shape)
print(y1.shape)
print()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression


tfidf = TfidfVectorizer(strip_accents=None, lowercase=False, preprocessor=None)
dtc_tfidf = Pipeline([('vect', tfidf), ('dtc', LogisticRegression())])
dtc_tfidf.fit(text, y1)
pred = dtc_tfidf.predict(test_text)
submission.cohesion = pred.astype(float)

dtc_tfidf.fit(text, y2)
pred = dtc_tfidf.predict(test_text)
submission.syntax = pred.astype(float)

dtc_tfidf.fit(text, y3)
pred = dtc_tfidf.predict(test_text)
submission.vocabulary = pred.astype(float)

dtc_tfidf.fit(text, y4)
pred = dtc_tfidf.predict(test_text)
submission.phraseology = pred.astype(float)

dtc_tfidf.fit(text, y5)
pred = dtc_tfidf.predict(test_text)
submission.grammar = pred.astype(float)

dtc_tfidf.fit(text, y6)
pred = dtc_tfidf.predict(test_text)
submission.conventions = pred.astype(float)
```

```python
submission.to_csv('./submission.csv')
print(submission)
```
